### 8. Etapa de Clustering

Para esta etapa del analisis probaremos diferentes tecnicas de clustering para agrupar nuestras instancias del dataset y luego elegiremos la que más se ajuste a las necesidades de los objetivos propuestos

#### K-Means Cluster

##### Buscamos la cantidad optima de cluster a través del método Codo de Jambú

```python
wcss = []
for i in range(1, 16):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(dataset_norm_cluster)
    wcss.append(kmeans.inertia_)
```

Graficamos los resultados de WCSS para formar el codo de Jambú

```python
pyplot.plot(range(1, 16), wcss)
pyplot.title('Codo de Jambú')
pyplot.xlabel('Número de Clusters')
pyplot.ylabel('WCSS')
pyplot.show()
```

Como podemos ver el valor de WCSS dejó de disminuir de manera drástica a partir del cluster 4, por lo tanto usameros la técnica de clustering de k.means con un valor de 4 clusters

```python
clusters_models['k-means'] = KMeans(n_clusters=4)
```

```python
k_means_clustering = clusters_models['k-means'].fit(dataset_norm_cluster)
```

Ahora procedemos a agregar los clusters a nuestro dataset

```python
kmeans_clusters = pd.DataFrame(k_means_clustering.labels_, columns=['cluster'])
kmeans_clusters
```

Ahora procedemos a graficar los clusters a través de la técnica de análisis de componentes principales (PCA)

```python
pca = PCA(n_components=2)
pca_dataset = pca.fit_transform(dataset_norm_cluster)
pca_df = pd.DataFrame(data = pca_dataset, columns= ['Componente_1', 'Componente_2'])
pca_clusters_dataset = pd.concat([pca_df, kmeans_clusters], axis=1)

pca_clusters_dataset
```

Procedemos a graficar

```python
fig = pyplot.figure(figsize = [6, 6])
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel('Componente_1', fontsize = 15)
ax.set_ylabel('Componente_2', fontsize = 15)
ax.set_title('Cluster K-Means', fontsize = 20)
color_theme = np.array(['green', 'yellow', 'purple', 'orange'])
ax.scatter(x = pca_clusters_dataset.Componente_1, y = pca_clusters_dataset.Componente_2,
          c=color_theme[pca_clusters_dataset.cluster], s=50)
pyplot.show()
```

```python
lr_pred_train_1 = regresiones_cluster['cluster_1'].predict(X_train_1)
print('Train Error: ' + str(np.sqrt(mean_squared_error(Y_train_1,lr_pred_train_1))))

lr_pred_1 = regresiones_cluster['cluster_1'].predict(X_validation_1)
print('Test Error: ' + str(np.sqrt(mean_squared_error(Y_validation_1,lr_pred_1))))
```
